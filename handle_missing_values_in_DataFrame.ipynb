```python
import pandas as pd
df = pd.read_csv("dataset/titanic.csv", usecols=['Age','Fare','Survived'])
df
```

```python
df.isnull().sum()
```

```python
df1 = df.fillna(0)
df1
```

```python
df1.isnull().sum()
```

```python
df2 = df.fillna(method="ffill")
df2
```

```python
df2.isnull().sum()
```

```python
df3 = df.fillna(method="ffill",limit=1)
df3
```

```python
df3.isnull().sum()
```

```python
df4 = df.fillna(method="bfill")
df4 
```

```python
df4.isnull().sum()
```

```python
df5 = df.interpolate() 
df5
```

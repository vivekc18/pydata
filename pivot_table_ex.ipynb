```python
import pandas as pd
import numpy as np
df=pd.read_csv("dataset/titanic.csv")
df
```

create a pivot table to diplay number of survival by gender and by class

```python
table=pd.pivot_table(df,index=['Pclass','Sex'],aggfunc='sum',values='Survived')
table
```

create pivot table and find survival rate by gender on various class

```python
table1=pd.pivot_table(df,index=['Sex'],columns='Pclass',values='Survived',aggfunc='count')
table1
```

create pivot table and find survival rate by gender

```python
table1=pd.pivot_table(df,index='Sex',values='Survived')
table1
```

create pivot table and find survival rate by gender and age wise of various class

```python
table1=pd.pivot_table(df,index=['Sex','Age'],columns='Pclass',values='Survived')
table1
```

create a pivot table and count survival by gender, catageroies wise age of various classes
Note: Age catagoeries (0,10),(10,30),(30,60),(60,80)

```python
age=pd.cut(df['Age'],[0,10,30,60,80])
age
```

```python
age=pd.pivot_table(df,index=['Sex',age],columns='Pclass',values='Survived',aggfunc='count')
age
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder,StandardScaler,MinMaxScaler,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('dataset/iris.csv')
df
```

```python
x = df.iloc[:,0:3]
```

```python
x
```

```python
y = df.iloc[:,-1]
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x , y , train_size=0.8 , random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
le = LabelEncoder()
y_tr = le.fit_transform(y_train)
```

```python
y_te = le.fit_transform(y_test)
```

```python
log = LogisticRegression()
```

```python
log.fit(x_train,y_tr)
```

```python
y_pred = log.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix
```

```python
print("Accuracy score =",accuracy_score(y_te,y_pred))
```

```python
pd.DataFrame(confusion_matrix(y_te , y_pred)) # if all the values are comes along the diagonal then accuracy is 100%
```

```python
result = pd.DataFrame()
result['Actual Label'] = y_test
result['Decision Tree Prediction'] = y_pred
```

```python
result
```

```python
from sklearn.metrics import precision_score
precision_score(y_te, y_pred, average='weighted')
```

```python
from sklearn.metrics import recall_score
recall_score(y_te, y_pred, average='micro')
```

```python
from sklearn.metrics import f1_score
f1_score(y_te, y_pred, average='micro')
```

```python
import pandas as pd 
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('dataset/bank-additional-full.csv')
df
```

```python
df.info()
```

```python
df.drop('duration',axis=1,inplace=True)
```

```python
df
```

```python
x=df.iloc[:,:-1]
x
```

```python
y = df.iloc[:,-1]
y
```

```python
x= pd.get_dummies(x,drop_first=True)
print(len(x.columns))
x.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.67,random_state=2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(x_train,y_train)
```

```python
a=dtc.predict(x_test)
a
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,a)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,a)
```

```python
from sklearn.metrics import recall_score
recall_score(y_test,a,average='micro')
```

```python
from sklearn.metrics import precision_score
precision_score(y_test,a,average='micro')
```

```python
from sklearn.metrics import precision_score
precision_score(y_test,a,average='macro')
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier(criterion='entropy',splitter='random',max_depth=10)
```

```python
dtc.fit(x_train,y_train)
```

```python
a=dtc.predict(x_test)
a
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,a)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,a)
```

```python
from sklearn.metrics import recall_score
recall_score(y_test,a,average='micro')
```

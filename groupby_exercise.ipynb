```python
#this will delete the column that is provided
df1=df.drop(["PassengerId","Name","Ticket","Cabin"],axis=1)
df1
```

```python
df1.describe()
```

```python
df1.isnull().sum()
```

```python
#this will fill the forward values of the missing value in the dataframe 
df2=df1.fillna(method="ffill")
df2
```

```python
df2.isnull().sum()
```

```python
#groupby returns the no of passengers survived or not,if 0.0 it will be non survived if 0.1 it will be survived
survived=df2.groupby('Survived')['Survived'].count()
survived
```

```python
#groupby returns the
passenger=df2.groupby('Pclass')['Pclass'].count()
passenger
```

```python
survived=df2.groupby('Sex')['Sex'].count()
survived
```

```python
survived=df2.groupby('Embarked')['Embarked'].count()
survived
```

```python
survived=df2.groupby('Sex')['Survived'].sum()
survived
```

```python
survived=df2.groupby(['Pclass','Sex','Survived'])['Survived'].count()
survived
```

```python
survived=df2.groupby(['Pclass'])['Survived'].sum()
survived
```

```python
survived=df2.groupby(['Pclass'])['Survived'].count()
survived
```

```python
survived=df2.groupby(['Pclass'])['Survived'].sum()/df2.groupby(['Pclass'])['Survived'].count()*100
survived
```

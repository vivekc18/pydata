```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('dataset/Breast_Cancer_Wisconsin.csv')
```

```python
df
```

```python
df.info()
```

```python
df['diagnosis'].unique()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
y=label_encoder.fit_transform(df['diagnosis'])
```

```python
y
```

```python
x=df.drop(['diagnosis','id'],axis=1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
```

```python
rfc.fit(x_train,y_train)
```

```python
y_pred=rfc.predict(x_test)
```

```python
rfc.score(x_train,y_train)
```

```python
rfc.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import precision_score
precision_score(y_test,y_pred,average='micro')
```

```python
n_estimators=[20,60,100,120,180]
min_samples_split=[0.15,0.25,0.30,0.40,0.70]
max_features=[0.2,0.4,0.6,0.7,1.0]
```

```python
param_grid={'n_estimators':n_estimators,
           'min_samples_split':min_samples_split,
           'max_features':max_features,
           }
print(param_grid)
```

```python
rf=RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
rf_grid=GridSearchCV(estimator=rf,
                    param_grid=param_grid,
                    verbose=2,
                    cv=5,
                    n_jobs=-1
                    )
```

```python
rf_grid.fit(x_train,y_train)
print(rf_grid.best_params_)
print(rf_grid.best_score_)
```

```python
rfc=RandomForestClassifier(max_features=0.2, min_samples_split=0.15, n_estimators=100)
rfc.fit(x_train,y_train)
rfc.score(x_train,y_train)
```

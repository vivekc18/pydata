```python
import pandas as pd
```

```python
df = pd.read_csv('dataset/iris.csv')
df
```

```python
df.isnull().sum()
```

```python
x = df.iloc[:,0:4]
x
```

```python
y = df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x , y , train_size=0.8 , random_state=2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_tr = le.fit_transform(y_train)
y_te = le.fit_transform(y_test)
```

```python
y_tr
```

```python
y_te
```

```python
from sklearn.svm import SVC 
classifier = SVC(kernel='linear', random_state=0)  
classifier.fit(x_train , y_tr)  
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix
print("Accuracy score =",accuracy_score(y_te,y_pred))
```

```python
pd.DataFrame(confusion_matrix(y_te , y_pred)) # if all the values are comes along the diagonal then accuracy is 100%
```

```python
result = pd.DataFrame()
result['Actual Label'] = y_test
result['Decision Tree Prediction'] = y_pred
```

```python
result
```

```python
from sklearn.metrics import precision_score
precision_score(y_te, y_pred, average='weighted')
```

```python
from sklearn.metrics import recall_score
recall_score(y_te, y_pred, average='micro')
```

```python
from sklearn.metrics import f1_score
f1_score(y_te, y_pred, average='micro')
```

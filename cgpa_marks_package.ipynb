```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("dataset/cgpa_marks_package.csv")
df
```

```python
df.info()
```

```python
x=df[["cgpa","Interview_marks"]]
x
```

```python
y=df[["package"]]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,train_size=0.8,random_state=2)
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
A=lr.predict(x_test)
```

```python
A
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test,A)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,A)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,A)
mse
```

```python
np.sqrt(mse)
```

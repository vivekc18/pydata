```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('dataset/placement.csv')
df
```

```python
df['placement_exam_marks'].describe()
```

```python
sns.boxplot(df['placement_exam_marks'])
```

```python
Q1=df['placement_exam_marks'].quantile(0.25)
Q1
```

```python
Q3=df['placement_exam_marks'].quantile(0.75)
Q3
```

```python
iqr=Q3-Q1
```

```python
iqr
```

```python
ul=Q3+1.5*iqr
ul
```

```python
ll=Q1-1.5*iqr
ll
```

```python
df[(df['placement_exam_marks']>ul) | (df['placement_exam_marks']<ll)]
```

```python
new_df=df.copy()
new_df
```

```python
new_df['placement_exam_marks']=np.where(new_df['placement_exam_marks']>ul,
                                        ul,
                                        np.where(new_df['placement_exam_marks']<ll,
                                                 ll,
                                                 new_df['placement_exam_marks']))
```

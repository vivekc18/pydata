## word tokenize

```python
from nltk.tokenize import sent_tokenize,word_tokenize
```

```python
txt="India is my country. All Indians are my brothers and sisters. I love my country and I am proud of its rich and varied heritage. "
```

```python
word_token=word_tokenize(txt)
print(word_token)
```

## sentence tokenize

```python
sent_token=sent_tokenize(txt)
print(sent_token)
```

## word distribution

```python
print(nltk.FreqDist(nltk.word_tokenize(txt)))
nltk.FreqDist(txt.split()).plot()
```

## word cloud

```python
from wordcloud import WordCloud
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
word="India is my country. All Indians are my brothers and sisters. I love my country and I am proud of its rich and varied heritage."
stopword=set(stopwords.words('english'))
wordcloud=WordCloud(background_color='white',stopwords=stopword,min_font_size=8)
out=wordcloud.generate(word)
fig=plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)
plt.imshow(out,interpolation='bilinear')
plt.axis('off')
plt.show()
```

## droping stopwords

```python
import nltk
from nltk.corpus import stopwords
stop_words=stopwords.words('english')
sent="India is a republic nation we are proud Indians . we love India . Jai hind"
word_tokens=word_tokenize(sent)
filtered_sentence=[w for w in word_tokens if w.lower() not in stop_words]
print(word_tokens)
print(filtered_sentence)
```

## droping panctuations

```python
import re
sent="India ,is, a, republic, nation, we, are, proud, Indian . we ,love, India . Jai, hind"
punch_remove=re.sub(r'[^\w\s]','',sent)
print(sent)
print(punch_remove)
```

```python
sent3="1947 was when India become independent"
print(re.search(r"[\w ]+",sent3))
```

## lemmatization

```python
from nltk.stem.wordnet import WordNetLemmatizer
lemmaobj=WordNetLemmatizer()
lemmaobj.lemmatize("went",pos='v')
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem.wordnet import WordNetLemmatizer
sent="went gone running ran playing played are is"
word_tokens=word_tokenize(sent)
lemma_vocab=[]
lemmaobj=WordNetLemmatizer()
for i in word_tokens:
    lemma_vocab.append(lemmaobj.lemmatize(i,pos='v'))
print(lemma_vocab)
```

## stemmer

```python
from nltk.stem.snowball import SnowballStemmer
stemobj=SnowballStemmer("english")
stemobj.stem("walked")
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer
sent=" His team are not winning"
word_tokens=word_tokenize(sent)
stemmed_vocab=[]
stemobj=SnowballStemmer("english")
for i in word_tokens:
    stemmed_vocab.append(stemobj.stem(i))
print(stemmed_vocab)
```
